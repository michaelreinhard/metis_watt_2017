# Basic plotting in matplotlib

This short notebook illustrates how to perform some helpful plotting functions.

## 1.  Loading and plotting a two-dimensional dataset

There are a variety of functions for loading in data e.g., numpy, pandas, etc.,.  Here we will use numpy's text loader since the point of this notebook is to explore plotting functionality.

First we load in a simple two-dimensional dataset.

```python
import numpy as np         # numerical linear algebra library - super useful

# load data
data = np.loadtxt('datasets/toy_regression_data.csv',delimiter=',')
```

We can examine the contents of *data* by placing it in a cell and activating it.  Alternatively you can say *print data* to do this.

```python
data
```

We can see how many rows and columns data has using numpy's shape command.  

```python
np.shape(data)
```

This says that the data array has 22 rows and 2 columns.

Since this constitutes a two-dimensional dataset, lets slice out each coordinate's data.  We do this in the next Python cell.

```python
x = data[:,0]
y = data[:,1]
```

Lets plot this data using matplotlib - a popular Python plotting library.  First we use the *scatter* functionality to plot the individual points.

```python
# activating the next line allows you to plot as many figures as you want inline in a jupyter notebook.
%matplotlib inline
```

```python
import matplotlib.pyplot as plt          # very common import statement for matplotlib
plt.scatter(x,y)
```

One can change all sorts of aspects of this figure - e.g., the size or color of the points, the placement / labels of axes and tickmarks, title, etc., - by studying the [matplotlib documentation](http://matplotlib.org/).  For example, in the next cell we again scatter plot the data but adjust the size and color of the points, label each axes, and add a title.

```python
plt.scatter(x,y,s = 50, color = 'r')
plt.xlabel('this is the x axis')
plt.ylabel('this is the y axis')
plt.title(' this is the title')
```

c an alternative to the scatter plot, we can simply 'plot' the data using matplotlib.  This function will connect each pair of points sequentially (as listed from top to bottom in the data array) using a line segment.

```python
plt.plot(x,y)
```

Again you can adjust the color and thickness of the plotting lines, axes labels, and title by using matplotlib functionality.

```python
plt.plot(x,y,linewidth = 5, color = 'r')
plt.xlabel('this is the x axis')
plt.ylabel('this is the y axis')
plt.title(' this is the title')
```
